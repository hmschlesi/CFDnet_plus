```python
import matplotlib.pyplot as plt
import numpy as np
import os
import re
```

```python
folders='ER600_Tin300_Uin040_TwallAdiabatic'
path_to_main='data/kolmogorv/'
```

```python
path_to_main='airfoil2d/'
folders='airFoil_300'
```

```python
subfolders=os.listdir(path_to_main+folders+'/VTK')
subfolders
```

```python
folder = []
files = []

for j in subfolders:
    #print(path_to_main+folders+'/VTK/'+j)
    if os.path.isdir(path_to_main+folders+'/VTK/'+j) == True: 
        folder.append(path_to_main+folders+'/VTK/'+j+'/')
        files.append(path_to_main+folders+'/VTK/'+j+'/'+'internal.vtu')
folder
```

```python
t=[]
for txt in folder:
    spl=txt.split('/')
    l=len(spl)
    ts=spl[l-2].replace(folders+'_','')
    #print(ts)
    t.append(ts)
t
```

```python
t_f=[int(i) for i in t]
#t_f.sort()
print(t_f)
```

```python
t_max=max(t_f)
print(t_max)
```

```python
idx = [j for j, item in enumerate(files) if re.search(str(int(t_max)), item)]
```
